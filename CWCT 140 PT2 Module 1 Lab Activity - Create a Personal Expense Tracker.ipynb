```python
import matplotlib.pyplot as plt
from fractions import Fraction
```

```python
expenses = [
{"Date":"2024-08-01", "Category":"Food", "Amount":12.50},
{"Date":"2024-08-02", "Category":"Transportation", "Amount":3.75},
{"Date":"2024-08-03", "Category":"Entertainment", "Amount":15.00},
{"Date":"2024-08-04", "Category":"Food", "Amount":8.25},
{"Date":"2024-08-05", "Category":"Transportation", "Amount":2.50},
{"Date":"2024-08-06", "Category":"Entertainment", "Amount":30.00}
]
```

Calculate Total

```python
total_expenses = 0.0
for expense in expenses:
    total_expenses += expense["Amount"]
```

```python
print(f"Total Expenses: ${total_expenses:.2f}")
```

Category Totals

```python
category_totals = {}
for expense in expenses:
    if expense["Category"] in category_totals:
        category_totals[expense["Category"]] += expense["Amount"]
    else:
        category_totals[expense["Category"]] = expense["Amount"]
```

Display Category Totals

```python
for category in category_totals: 
    print(f"{category}: ${category_totals[category]:.2f}")
```

```python
categories = list(category_totals.keys())
totals = list(category_totals.values())
```

```python
plt.bar(categories, totals)
plt.xlabel('Category')
plt.ylabel('Total Amount ($)')
plt.title('Total Expense by Category')
```

```python
plt.pie(totals, labels=categories, autopct= '%1.1f%%')
plt.title('Spending Breakdown by Category')
```

Calculate the fraction of the total for each category

```python
for category in category_totals:
    fraction = category_totals[category]/total_expenses
    print(f"fraction for {category}: {fraction:.2%} of total expenses")
```

Display the results as a numeric rational fraction

```python
for category in category_totals:
    fraction = Fraction(int(category_totals[category]*100), int(total_expenses*100))
    print(f"Fraction for {category}: {fraction} of total expense")
```

Based on data analytics, we found that entertainment accounts for a majority of funds spent.
